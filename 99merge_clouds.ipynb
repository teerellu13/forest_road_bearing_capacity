```python
import os
import numpy as np
import laspy

folder_path = r"C:\Users\telukkari\Documents\Python\forest_road_bearing_capacity\data\bc_measurements"
output_path = r"C:\Users\telukkari\Documents\Python\forest_road_bearing_capacity\data\bc_measurements\all_measurements.las"

all_points = []
header_template = None

for filename in os.listdir(folder_path):
    if not filename.lower().endswith((".las", ".laz")):
        continue

    file_path = os.path.join(folder_path, filename)
    las = laspy.read(file_path)

    if header_template is None:
        header_template = las.header

    # convert to numpy structured array (safe)
    pts = np.vstack((las.x, las.y, las.z, las.bearing_capacity)).T

    all_points.append(pts)

# --- Merge ---
print("Merging", len(all_points), "files...")
merged_points = np.concatenate(all_points)

# --- Write final LAS ---
merged_las = laspy.LasData(header_template)
merged_las.x = merged_points[:,0]
merged_las.y = merged_points[:,1]
merged_las.z = merged_points[:,2]
merged_las.bearing_capacity = merged_points[:,3]
merged_las.write(output_path)

print("Merged LAS saved to:", output_path)
```
